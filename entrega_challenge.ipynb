### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion = sum(tienda['Precio'])
print('El ingreso total estimado de la tienda 1 es de: $'+ str(facturacion))
facturacion2 = sum(tienda2['Precio'])
print('El ingreso total estimado de la tienda 2 es de: $'+ str(facturacion2))
facturacion3 = sum(tienda3['Precio'])
print(f'El ingreso total estimado de la tienda 3 es de: ${facturacion3}')
facturacion4 = sum(tienda4['Precio'])
print(f'El ingreso total estimado de la tienda 4 es de: ${facturacion4}')
```

# 2. Ventas por categoría

```python
ventaxcategori = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print('Ventas por categoría del producto de la tienda1:\n', ventaxcategori)

ventaxcategori2 = tienda2.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
print('Ventas por categoría del producto de la tienda1:\n',ventaxcategori2 )

ventaxcategori3 = tienda3.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
print('Ventas por categoría del producto de la tienda1:\n',ventaxcategori3 )

ventaxcategori4 = tienda4.groupby(['Categoría del Producto']).size().sort_values(ascending=False)
print('Ventas por categoría del producto de la tienda1:\n',ventaxcategori4 )
```

# 3. Calificación promedio de la tienda


```python
calificacionmedia = tienda['Calificación'].mean()
print(f'La calificación promedio de la tienda 1 es de: {calificacionmedia.round(2)}')

calificacionmedia2 = tienda2['Calificación'].mean()
print(f'La calificación promedio de la tienda 2 es de: {calificacionmedia2.round(2)}')

calificacionmedia3 = tienda3['Calificación'].mean()
print(f'La calificación promedio de la tienda 3 es de: {calificacionmedia3.round(2)}')

calificacionmedia4 = tienda4['Calificación'].mean()
print(f'La calificación promedio de la tienda 4 es de: {calificacionmedia4.round(2)}')
```

# 4. Productos más y menos vendidos

```python
productosmasvendidos = tienda.groupby('Producto').size().sort_values(ascending=False)
print('Productos más vendidos de la tienda 1:\n', productosmasvendidos)

productosmenosvendidos = tienda.groupby('Producto').size().sort_values(ascending=True)
print('Productos menos vendidos de la tienda 1:\n', productosmenosvendidos)

productosmasvendidos2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
print('Productos más vendidos de la tienda 2:\n', productosmasvendidos2)

productosmenosvendidos2 = tienda2.groupby('Producto').size().sort_values(ascending=True)
print('Productos menos vendidos de la tienda 2:\n', productosmenosvendidos2)
productosmasvendidos3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
print('Productos más vendidos de la tienda 3:\n', productosmasvendidos3)

productosmenosvendidos3 = tienda3.groupby('Producto').size().sort_values(ascending=True)
print('Productos menos vendidos de la tienda 3:\n', productosmenosvendidos3)
productosmasvendidos4 = tienda4.groupby('Producto').size().sort_values(ascending=False)
print('Productos más vendidos de la tienda 4:\n', productosmasvendidos4)

productosmenosvendidos4 = tienda4.groupby('Producto').size().sort_values(ascending=True)
print('Productos menos vendidos de la tienda 4:\n', productosmenosvendidos4)
```

# 5. Envío promedio por tienda

```python
valordeenviotienda1 = tienda['Costo de envío'].mean()
print(f'El valor del envío promedio de la tienda 1 es de: ${valordeenviotienda1.round(2)}')

valordeenviotienda2 = tienda2['Costo de envío'].mean()
print(f'El valor del envío promedio de la tienda 2 es de: ${valordeenviotienda2.round(2)}')

valordeenviotienda3 = tienda3['Costo de envío'].mean()
print(f'El valor del envío promedio de la tienda 3 es de: ${valordeenviotienda3.round(2)}')

valordeenviotienda4 = tienda4['Costo de envío'].mean()
print(f'El valor del envío promedio de la tienda 4 es de: ${valordeenviotienda4.round(2)}')
```
